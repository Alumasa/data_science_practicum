```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'a': [2, 3, 4, 5], 'b': [4, 9, 16, 25]})
print(df)
df.plot();
```

```python
df.plot(title='A and B')
```

```python
plt.title('A and b')
```

```python
df.plot(style='x')
```

```python
df.plot(style='o-') # 'o-' a circle and a line
```

```python
df.plot(style=':') # 'o-' a circle and a line
```

```python
df.plot(x='b', y='a', style='o-', xlim=(0, 30), grid=True);
```

```python
# building a small graph
df.plot(x='b', y='a', style='o-', xlim=(0, 30), grid=True, figsize=(1, 3))

# building a large graph
df.plot(x='b', y='a', style='o-', xlim=(0, 30), grid=True, figsize=(10, 3));
```
